```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

from astropy.table import QTable

from astropy import units as u
from astropy import constants as const
```

```python
R = np.arange(1,101,1) / 10 * (6378136 *u.m)
M = np.arange(1,101,1) / 10 * u.earthMass
```

```python
EarthR = u.def_unit('EarthR', 6378136 *u.m)
```

```python
density = M / (4/3 * np.pi * R**3)
```

```python
density.to(u.g/u.cm**3)
```

```python
T = QTable()
```

```python
T['M'] = M
T['R'] = R.to(EarthR)
T['Density'] = density.to(u.g/u.cm**3)
```

```python
x = 0.5 * (u.g/u.cm**3)
y = 5.0 * (u.g/u.cm**3)
mask = np.where( (T['Density'] > x) & (T['Density'] < y))
```
